```python
import ee
from google.colab import drive

drive.mount('/content/drive')

ee.Authenticate()

ee.Initialize(project='peaceful-field-489706-m3')
print("Xác thực Earth Engine Thành Công!")
```

```python
MY_FOLDER = '/content/drive/MyDrive/SmartSite/Step 1: Extract/Results_CSV'
FILE_POSTFIX = 'VIIRS'

regions = {
    'DaNang': ee.Geometry.Rectangle([108.0600, 15.9200, 108.3200, 16.1700]),
    'HaNoi': ee.Geometry.Rectangle([105.2800, 20.5500, 106.0200, 21.4000]),
    'HCM': ee.Geometry.Rectangle([106.3500, 10.3500, 106.8500, 11.1600])
}

dataset = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG") \
            .filterDate('2023-01-01', '2023-12-31') \
            .select('avg_rad') \
            .median()

print(f"Đang gửi yêu cầu tải ảnh vệ tinh vào thư mục: '{MY_FOLDER}'...")

for city_name, geometry in regions.items():
    custom_filename = f"{city_name}_{FILE_POSTFIX}"

    task = ee.batch.Export.image.toDrive(
        image=dataset,
        description=custom_filename,
        folder=MY_FOLDER,
        fileNamePrefix=custom_filename,
        region=geometry,
        scale=463.83,
        crs='EPSG:4326',
        maxPixels=1e10
    )
    task.start()
    print(f"Đã ra lệnh xuất file: {custom_filename}.tif")
```
